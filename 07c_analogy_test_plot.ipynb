```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from collections import defaultdict
import operator

pd.set_option('display.notebook_repr_html', False)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 150)
pd.set_option('display.max_seq_items', None)
 
%matplotlib inline  

import seaborn as sns
sns.set(font_scale=1.5)
sns.set_style("whitegrid")

from matplotlib.backends.backend_pdf import PdfPages
```

```python
df = pd.read_table("./analogy_test_result.txt")
```

```python
df.head()
```

```python
df_sub = df.loc[df['category'] == "total"]
g = sns.factorplot(x="w2v_dim", y="accuracy", hue="subreddit", col="w2v_model", 
                   data=df_sub, kind="bar",
                   size=4, aspect=.7)

g.savefig('plot/analog_test.pdf')
```

```python
w2v_model = "skipgram"
df_by_w2vmodel = df.loc[df['w2v_model'] == w2v_model]
g = sns.factorplot(x="w2v_dim", y="accuracy", hue="subreddit", col="category",
                       data=df_by_w2vmodel, kind="bar", col_wrap=3, size=5, aspect=.7)
```

```python
w2v_model = "cbow"
df_by_w2vmodel = df.loc[df['w2v_model'] == w2v_model]
g = sns.factorplot(x="w2v_dim", y="accuracy", hue="subreddit", col="category",
                       data=df_by_w2vmodel, kind="bar", col_wrap=3, size=5, aspect=.7)
```
